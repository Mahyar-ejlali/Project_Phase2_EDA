## Shortcomings


There are some feature cannot be captured due to lack of data:

**Mileage data:** The mileage on component at repair event is not collected. This could give us a very good insight about failure rate and expected amount of labor hours.

**Cost profile data:** The amount of money has been spent on each car is not included. This could be a good indicator for failure rate and  expected amount of labor hours.

**Loading profile data**: Type of commodities were transfered by each railcar is not provided. It could be another way to evaluate the required amount of labor hours.
